# What this file does 

I got Age by Sex by Ambulatory Disability estimates from Factfinder. (source: https://factfinder.census.gov/faces/nav/jsf/pages/index.xhtml, table: B18105 "Age by Sex by Ambulatory Disability", 2017 5-year estimates) I combined male and female estimates by census tract, wrangled the data for a join and finally joined everything in a QJIS analysis. (ambulatory-disability)

Major caveat: The maps don't represent the margins of error that accompany estimates of males and females between 5-17 with ambulatory disabilities. Use with caution.

```python
df2 = pd.read_csv('data-cleaned-ambulatory.csv')
```

```python
def county_code(x):
    if x ==' Bronx County':
        return 2
    if x ==' New York County':
        return 1
    if x ==' Queens County':
        return 4
    if x ==' Kings County': 
        return 3
    if x ==' Richmond County':
        return 5
```

```python
df2['County_code'] = df2['County'].apply(lambda x: county_code(x))
```

```python
def numLen(num):
    return len(str(round(num)))
```

```python
numLen(113.50)
```

```python
def moving_zeroes(x):

    if numLen(x)==4:
        return int(x*100)
    
    if  numLen(x)==3:
        return '0'+ str(int(x*100))
    
    if  numLen(x)==2:
        return '00'+ str(int(x*100))
    
    if numLen(x)==1:
        return '000'+ str(int(x*100))
```

```python
#test cases 

print(moving_zeroes(1145))
print(moving_zeroes(1145.40))

print(moving_zeroes(113))
print(moving_zeroes(114.40))
print(moving_zeroes(114.45))
print(moving_zeroes(28.45))
print(moving_zeroes(55))
print(moving_zeroes(344))
print(moving_zeroes(1))
```

```python
df2['adjusted_zeroes'] = df2['tract_no'].apply(lambda x: moving_zeroes(x))
```

```python
df2[['adjusted_zeroes', 'tract_no']]
```

```python
df2['boro_ct201'] = df2['County_code'].astype(str) + df2['adjusted_zeroes'].astype(str)
```
